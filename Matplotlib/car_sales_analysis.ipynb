```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Data/Car_sales.csv')
```

```python
most_sold_manufacturer = data['Manufacturer'].value_counts()
most_sold_manufacturer.idxmax()
```

```python
plt.scatter(data['Engine_size'], data['Horsepower'])
plt.title('Scatter Plot: Engine Size vs Horsepower')
plt.xlabel('Engine size')
plt.ylabel('Horsepower')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x='Manufacturer', y='Horsepower', data=data,errorbar=None)
plt.title('Average Horsepower by Manufacturer')
plt.xticks(rotation=45, ha='left')
plt.xlabel('Manufacturer')
plt.ylabel('Average Horsepower')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Vehicle_type', y='Price_in_thousands', data=data)
plt.xticks(rotation=30, ha = 'right')
plt.title('Box Plot: Price by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Price_in_thousands')
plt.show()
```
